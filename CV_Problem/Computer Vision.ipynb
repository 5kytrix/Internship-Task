# Converting labels to one hot encoded form

```python
from keras.utils import to_categorical
label = to_categorical(label)
```

# Train, Test and Validation Split

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(training, label, test_size=0.2, random_state=13)
```

```python
train_X,valid_X,train_label,valid_label = train_test_split(train_X, train_y, test_size=0.2, random_state=13)
```

```python
train_X = train_X.reshape((train_X.shape[0],28,28,1))
```

# Using Image Data Generator and applying zca whitening to training data for Data Augmentation

```python
from keras.preprocessing.image import ImageDataGenerator
from keras import backend as K
shift = 0.2
datagen = ImageDataGenerator(zca_whitening=True)
# fit parameters from data
datagen.fit(train_X)
print(train_X.shape)
# configure batch size and retrieve one batch of images
i=0
for X_batch, y_batch in datagen.flow(train_X, train_label):
    # create a grid of 3x3 images
    train_X = np.vstack((train_X,X_batch))
    train_label = np.vstack((train_label,y_batch))
    print(i)
    i = i+1
    if i==10:
        break
```

```python
train_X.shape
```

```python
import keras
from keras.models import Sequential,Input,Model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU
from keras.callbacks import ReduceLROnPlateau
from keras.optimizers import RMSprop
from keras.optimizers import Adam
from keras.callbacks import TensorBoard


batch_size = 64
epochs = 25
num_classes = 7
```

# CNN Model for training

```python
model = Sequential()
model.add(Conv2D(32,(3, 3), input_shape = (28,28,1), activation='relu'))
model.add(Conv2D(32,(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64,(2, 2), activation='relu'))
model.add(Conv2D(64,(2, 2), activation='relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(activation = 'relu', units=16))
model.add(Dense(activation = 'softmax', units=7))
model.summary()
model.compile(loss ='categorical_crossentropy', optimizer=Adam(lr=0.001),metrics =['accuracy'])
```

```python
valid_X = valid_X.reshape((valid_X.shape[0],28,28,1))
test_X = test_X.reshape((test_X.shape[0],28,28,1))
```

# Training the model

```python
train = model.fit(train_X, train_label, batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label))
```

# Test Accuracy of 86%
# Validation Acuracy of 85%

```python
test_eval = model.evaluate(test_X, test_y, verbose=0)
print('Test loss:', test_eval[0])
print('Test accuracy:', test_eval[1])
```
