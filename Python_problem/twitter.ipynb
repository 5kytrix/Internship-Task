```python
#importing all libraries
import tweepy
import numpy as np
import csv
import json
  
#Load twitter API after authorization of user
consumer_key = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX" 
consumer_secret = "XXXXXXXXXXXXXXXXXXXXXXXXXXXX"
access_key = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
access_secret = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
  
# Function to extract tweets 
def get_tweets(username): 
          
        # Authorization to consumer key and consumer secret 
        auth = tweepy.OAuthHandler(consumer_key, consumer_secret) 
  
        # Access to user's access key and access secret 
        auth.set_access_token(access_key, access_secret) 
  
        # Calling api 
        api = tweepy.API(auth) 
  
        #initialize a list to hold all the tweepy Tweets
        alltweets = [] 
        #make initial request for most recent tweets
        tweets = api.user_timeline(screen_name=username)
        #save most recent tweets
        alltweets.extend(tweets)
        #save the id of the oldest tweet less one
        oldest = alltweets[-1].id - 1
        #keep grabbing tweets until there are no tweets left to grab
        while len(tweets) > 0:
            #all subsiquent requests use the max_id param to prevent duplicates
            tweets = api.user_timeline(screen_name=username,count=200,max_id=oldest)
            #save most recent tweets
            alltweets.extend(tweets)
            #update the id of the oldest tweet less one
            oldest = alltweets[-1].id - 1

  
        #Dump the responses into JSON file
        print("Writing tweet objects to JSON please wait...")
        with open('tweet.json', 'a') as f:
            for tweet in alltweets:
                json.dump(tweet._json, f)
                #Separate each tweet so as to read properly as json.loads takes only a sngle entry
                f.write('\n')
  
# Driver code 
if __name__ == '__main__': 

    get_tweets("midasIIITD")
    
```

```python
#Initialize a list to hold the responses
tweets = []
for line in open('tweet.json', 'r'):
    #calling json.load to append each tweet separately
    tweets.append(json.loads(line))
```

```python
#printing the responses
tweets
```

```python
import pandas as pd
df = pd.DataFrame()
 
df['text'] = list(map(lambda tweet: tweet['text'], tweets))
 
df['date_time'] = list(map(lambda tweet: tweet['created_at'], tweets))
 
df['favorites'] = list(map(lambda tweet: tweet['favorite_count'], tweets))
 
df['retweet'] = list(map(lambda tweet: tweet['retweet_count'], tweets))
```
